- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets
from sklearn.cluster import KMeans
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
clu1 = KMeans(n_clusters=2)
clu2 = KMeans(n_clusters=3)
clu3 = KMeans(n_clusters=4)

clu1.fit(X)
clu2.fit(X)
clu3.fit(X)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
clu1.labels_
clu2.labels_
clu3.labels_
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=clu1.labels_)
print("sperate in 2 group")
plt.show()
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=clu2.labels_)
print("sperate in 3 group")
plt.show()
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=clu3.labels_)
print("sperate in 4 group")
plt.show()
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y)
print("ground true")
plt.show()
```
