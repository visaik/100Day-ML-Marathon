使用function y=(x+5)² 來練習學習率( leanrning rate ) 調整所造成的影響

# 作業目標:
    請嘗試使用不同的組合驗證 learning rate 對所需 iteration 的影響
    lr = [0.1, 0.0001]
    主要驗證 Lr 對於grandient 收斂的速度
    

```python
cur_x = 3 # The algorithm starts at x=3
precision = 0.000001 #This tells us when to stop the algorithm
max_iters = 10000 # maximum number of iterations
df = lambda x: 2*(x+5) #Gradient of our function 
```

```python
Learning_rate = 0.1
iters= 1
previous_step_size = 1

x_history_0 = [cur_x]
while previous_step_size > precision and iters < max_iters:
    print("the learning rate is {}, Iteration {}, X value is{}".format(Learning_rate, iters, cur_x))  
    previous_x = cur_x
    cur_x = previous_x - Learning_rate * df(previous_x)
    previous_step_size = abs(cur_x - previous_x)
    iters+=1
    x_history_0.append(cur_x)
```

```python
cur_x = 3
Learning_rate = 0.0001
iters= 1
previous_step_size = 1

x_history_1 = [cur_x]
while previous_step_size > precision and iters < max_iters:
    print("the learning rate is {}, Iteration {}, X value is{}".format(Learning_rate, iters, cur_x))  
    previous_x = cur_x
    cur_x = previous_x - Learning_rate * df(previous_x)
    previous_step_size = abs(cur_x - previous_x)
    iters+=1
    x_history_1.append(cur_x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(np.linspace(1,10000,10000),x_history_0, color='black')
plt.plot(np.linspace(1,10000,10000),x_history_1, color='red')
plt.show()
```
