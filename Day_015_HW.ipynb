```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# 設定 data_path
dir_data = './data/'
```

```python
# 讀取資料檔
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
app_train.shape
```

```python
app_train.head()
```

```python
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# 檢查每一個 column
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder, 以加入相關係數檢查
            app_train[col] = le.fit_transform(app_train[col])            
print(app_train.shape)
app_train.head()
```

```python
# 受雇日數為異常值的資料, 另外設一個欄位記錄, 並將異常的日數轉成空值 (np.nan)
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
app_train['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)

# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
```

### 相關係數
一樣，pandas 很貼心地讓我們可以非常容易計算相關係數

```python
# 觀察相關係數
Target_corr = app_train.corr()['TARGET'].copy()
print(Target_corr)
```

```python
#Target_corr.values.sort()
Target_corr.sort_values(inplace=True)
print("The top 15th corr")
print(Target_corr.tail(15))
print("The last 15th corr")
print(Target_corr.head(15))
```

Let's check Target and DAYS_EMPLOYED which is the largest corr except Target itself.

```python
#set the subset of raw
sub_set_1 = (app_train['TARGET'] == 1) & (~app_train['DAYS_EMPLOYED'].isnull())
sub_set_0 = (app_train['TARGET'] == 0) & (~app_train['DAYS_EMPLOYED'].isnull())
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.boxplot([app_train[sub_set_1]['DAYS_EMPLOYED'],app_train[sub_set_0]['DAYS_EMPLOYED']],labels=['Target=1', 'Target=0'])
axes.set_ylim(bottom=-7500)
fig.show()
```
