### 用 digits (dataset.load_digits()) 資料嘗試跑 t-SNE

```python
from sklearn import datasets

digits = datasets.load_digits()
X = digits.data
y = digits.target
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import manifold
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
tsne = manifold.TSNE(n_components=2, random_state=0, init='pca', learning_rate=200., early_exaggeration=12.)
X_tsne = tsne.fit_transform(X)
X_tsne = pd.DataFrame(X_tsne)
```

```python
fig, axe = plt.subplots(figsize=(10,10))
palette = sns.color_palette("bright", 10)
sns.scatterplot(x=0,y=1,hue=y,data=X_tsne,ax=axe, palette=palette, legend='full')
axe.set_title("t-SNE")
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca = pd.DataFrame(X_pca)
```

```python
fig, axe1 = plt.subplots(figsize=(10,10))
sns.scatterplot(x=0,y=1,hue=y,data=X_pca, ax=axe1,palette="Accent", legend='full')
axe1.set_title("PCA")
plt.show()
```
